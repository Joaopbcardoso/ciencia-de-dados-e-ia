```python
soma = 0
receita = 0
custo = 0
ROI = 0

while True:
    receita = float(input("Qual o valor da receita? "))
    custo = float(input("Qual o custo? "))
    ROI = ((receita - custo) / custo) * 100
    print(f"O retorno sobre o investimento é de {ROI:.2f}%")
    soma += ROI

    continuar = input("Deseja continuar calculando ROI? (s/n): ").upper()
    if continuar != "S" and continuar != "N":
        print("Resposta inválida. Digite apenas 'S' ou 'N'.")
        continuar = input("Deseja continuar calculando ROI? (s/n): ").upper()

    if continuar == "S":
        continue
    else:
        ver_soma = input("Deseja ver a soma dos ROI? (s/n): ").upper()
        if ver_soma != "S" and ver_soma != "N":
            print("Resposta inválida. Digite apenas 'S' ou 'N'.")
            ver_soma = input("Deseja ver a soma dos ROI? (s/n): ").upper()

        if ver_soma == "S":
            print(f"A soma dos ROIs foi {soma:.2f}")
        else:
            print("Fim da sessão")
        break
```

```python
#Criar Lista
cidades = []
#Definir um contador de cidades
contador_cidades = 1

#Formar um while para não repetir trÊs vezes o input.
while contador_cidades <= 3: # condição para repetir 3 vezes o input
        cidade = input(f"Qual a {contador_cidades}º ciadde")
        cidades.append(cidade)
        contador_cidades += 1
print(cidades)
```
