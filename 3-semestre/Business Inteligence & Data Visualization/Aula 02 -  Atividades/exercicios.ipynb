O exemplo abaixo ilustra como calcular a "Eficiência no Pagamento": a relação entre o valor pago e o valor empenhado(reservado) para a emenda

```python
import pandas as pd
import numpy as np

df = pd.read_csv("EmendasParlamentares.csv",encoding="iso-8859-1",sep=';')
#Tentamos converter diretamente. Importante, os valores fracionados na planilha estão usando , como separador e pandas espera .
df['Valor Empenhado'] = df['Valor Empenhado'].str.replace(',','.')
df['Valor Pago'] = df['Valor Pago'].str.replace(',','.')
df['Valor Empenhado'] = pd.to_numeric(df['Valor Empenhado'])
df['Valor Pago'] = pd.to_numeric(df['Valor Pago'])

# Exemplo de criação de atributo (Feature Engineering)
df['Eficiencia_Pagamento'] = (df['Valor Pago'] / df['Valor Empenhado']) * 100

# Ranking dos autores mais eficientes em liberar recursos
ranking = df.groupby('Nome do Autor da Emenda')['Eficiencia_Pagamento'].mean().sort_values(ascending=False)
ranking
```

# Exercício 1

```python
# Agrupa por ano e soma a coluna 'Valor Pago'
df['Valor Pago'] = pd.to_numeric(df['Valor Pago'].astype(str).str.replace(',', '.'), errors='coerce').fillna(0)

# Agrupamento por ano
total_pago_ano = df.groupby('Ano da Emenda')['Valor Pago'].sum().reset_index()

print("--- Valor Total Pago por Ano ---")
print(total_pago_ano.to_string(index=False))
```

# Exercício 2

```python
df = pd.read_csv("EmendasParlamentares.csv", encoding="iso-8859-1", sep=';')

df['Valor Empenhado'] = pd.to_numeric(df['Valor Empenhado'].astype(str).str.replace(',', '.'), errors='coerce').fillna(0)

# Cálculo de múltiplas métricas simultâneas
estatisticas = df.groupby('Região')['Valor Empenhado'].agg(['mean', 'std']).rename(columns={'mean': 'Média', 'std': 'Desvio Padrão'})

print("--- Estatísticas por Região ---")
print(estatisticas)
```

# Exercício 3

```python
df = pd.read_csv("EmendasParlamentares.csv", encoding="iso-8859-1", sep=';')

df['Valor Empenhado'] = pd.to_numeric(df['Valor Empenhado'].astype(str).str.replace(',', '.'), errors='coerce').fillna(0)

# nlargest(10) retorna os 10 maiores valores de forma eficiente
top_10 = df.groupby('Nome do Autor da Emenda')['Valor Empenhado'].sum().nlargest(10)

print("--- Top 10 Autores (Maior Valor Empenhado) ---")
print(top_10)
```
